```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_filepath = 'Resources/purchase_data.csv'

data_df = pd.read_csv(csv_filepath)
data_df.head()
```

```python
## player count
# print('### Player Count')
# player_count = (data_df['SN'].count())
# player_count
#leave it as '.unique' or change to '.count'
players_dic = {
    'Tot # Players': [len(data_df['SN'].unique())]
}

players_df = pd.DataFrame(players_dic)
players_df

# player_count2 = data_df["SN"].value_counts()
# player_count2
```

```python
### Purchasing Analysis (Total)
print('### Purchasing Analysis')
uni_items = data_df['Item Name'].count()
avg_price = data_df['Price'].mean()
tot_purchase = data_df['Price'].count()
tot_revenue = data_df['Price'].sum()

analysis_dic = {
    'Num Unique Items': [data_df['Item Name'].count()],
    'Avg. Price': [data_df['Price'].mean()],
    'Num of Purchases': [data_df['Price'].count()],
    'Tot Revenue': [data_df['Price'].sum()]
}



analysis_df = pd.DataFrame(analysis_dic)
analysis_df['Avg. Price'] = analysis_df['Avg. Price'].map('${:.2f}'.format)
analysis_df['Tot Revenue'] = analysis_df['Tot Revenue'].map('${:.2f}'.format)
analysis_df
```

```python
### Gender Demographics
print('### Gender Demographics')

## Percentage and Count of Male Players
## Percentage and Count of Female Players
## Percentage and Count of Other / Non-Disclosed
players_by_gender = data_df['Gender'].value_counts()
players_by_gender_df = pd.DataFrame(players_by_gender)

percent_players = (data_df['Gender'].value_counts() / (data_df['Gender'].count())) * 100
players_by_gender_df['Percent of Players'] = percent_players
gender_demos_df = pd.DataFrame(players_by_gender_df)
gender_demos_df

new_gender_demos_df = gender_demos_df.rename(columns = {
    'Gender': 'Total Count',
})

new_gender_demos_df['Percent of Players'] = new_gender_demos_df['Percent of Players'].map('{:.2f}%'.format)
new_gender_demos_df
```

```python
###Purchase Analysis (Gender)
# The below each broken by gender
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Gender

gender_df = data_df.groupby('Gender')

player_count = gender_df['SN'].count()
player_count_df = pd.DataFrame(player_count)
# player_count_df

avg_price = data_df['Price'].mean()
#avg_price_df = pd.DataFrame(avg_price)

tot_gender_pur = gender_df['Price'].sum()

norm_totals = tot_gender_pur / player_count

player_count_df['Average Pur Price'] = avg_price
gender_pur_df = pd.DataFrame(player_count_df)

new_gender_pur_df = gender_pur_df.rename(columns = {
    'SN': 'Purchase Count',
})

new_gender_pur_df['Total Purchase Price'] = tot_gender_pur
new_gender_pur2_df = pd.DataFrame(new_gender_pur_df)

new_gender_pur2_df['Avg. Pur Total per Person'] = norm_totals
tot_gender_pur_df = pd.DataFrame(new_gender_pur2_df)

tot_gender_pur_df['Average Pur Price'] = tot_gender_pur_df['Average Pur Price'].map('${:.2f}'.format)
tot_gender_pur_df['Total Purchase Price'] = tot_gender_pur_df['Total Purchase Price'].map('${:.2f}'.format)
tot_gender_pur_df['Avg. Pur Total per Person'] = tot_gender_pur_df['Avg. Pur Total per Person'].map('${:.2f}'.format)
tot_gender_pur_df
```

```python
# Age Demographics
# The below each broken into bins of 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Age Group

bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', 'Over 40']

binner_df = data_df.copy()

binner_df["Age Groups"] = pd.cut(binner_df['Age'], bins, labels = groups)
group_bin = binner_df.groupby(['Age Groups'])

binnercount = group_bin['SN'].count()
counttotal = data_df['SN'].count()
percentage = (binnercount / counttotal) * 100
percentage

Age_Perc = pd.DataFrame({
'Total Count': binnercount,
'Percentage of Players': percentage})

Age_Perc['Percentage of Players'] = Age_Perc['Percentage of Players'].map("{:.2f}%".format)
Age_Perc.head(10)
```

```python
# Age Demographics
# The below each broken into bins of 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Age Group
bins = [0,10,15,20,25,30,35,40,200]
groups = ['Under 10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', 'Over 40']

binned_df = data_df.copy()
binned_df['Age Groups'] = pd.cut(binned_df['Age'], bins, labels = groups)
binnedColumn = pd.cut(binned_df['Age'], bins, labels = groups)
grouped_bin = binned_df.groupby(['Age Groups'])

binnedcount = grouped_bin['Age'].count()
binnedavg = grouped_bin['Price'].mean()
binnedtotal = grouped_bin['Price'].sum()

binnedduplicate = data_df.drop_duplicates(subset='SN', keep='first')
binnedduplicate['Age Groups'] = pd.cut(binnedduplicate['Age'], bins, labels = groups)
binnedduplicate = binnedduplicate.groupby(['Age Groups'])

binnedavgperperson = (grouped_bin['Price'].sum() / binnedduplicate['SN'].count())
binnedavgperperson

age_demos = pd.DataFrame({"Purchase Count": binnedcount,
                         "Average Purchase Price": binnedavg,
                         "Total Purchase Value": binnedtotal,
                         "Avg. Pur Total per Person": binnedavgperperson})

age_demos['Average Purchase Price'] = age_demos['Average Purchase Price'].map("${:.2f}".format)
age_demos['Total Purchase Value'] = age_demos['Total Purchase Value'].map("${:.2f}".format)
age_demos['Avg. Pur Total per Person'] = age_demos['Avg. Pur Total per Person'].map("${:.2f}".format)
age_demos = age_demos[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg. Pur Total per Person']]

age_demos.head(10)
```

```python
#top spenders
sn_groups = data_df.groupby(['SN'])
sn_group_count = sn_groups['Item ID'].count()
sn_group_total = sn_groups['Price'].sum()
sn_group_avg = (sn_group_total / sn_group_count)

sn_demos = pd.DataFrame({'Purchase Count': sn_group_count,
                       'Average Purchase Price': sn_group_avg,
                       'Total Purchase Value': sn_group_total})

sn_demos = sn_demos.sort_values('Total Purchase Value', ascending = False)
sn_demos['Average Purchase Price'] = sn_demos['Average Purchase Price'].map("${:.2f}".format)
sn_demos['Total Purchase Value'] = sn_demos['Total Purchase Value'].map("${:.2f}".format)
sn_demos = sn_demos[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
sn_demos.head()
```

```python
#most popular items
grouped_item = data_df.groupby(['Item ID', 'Item Name'])
grouped_item_count = grouped_item['SN'].count()
grouped_price_sum = grouped_item['Price'].sum()
grouped_item_price = (grouped_price_sum / grouped_item_count)
grouped_item_value = (grouped_item_price * grouped_item_count)

popular_item = pd.DataFrame({'Purchase Count': grouped_item_count,
                        'Item Price': grouped_item_price,
                        'Total Purchase Value': grouped_item_value})

popular_item = popular_item.sort_values('Purchase Count', ascending = False)

popular_item['Item Price'] = popular_item['Item Price'].map('${:.2f}'.format)
popular_item['Total Purchase Value'] = popular_item['Total Purchase Value'].map('${:.2f}'.format)
popular_item = popular_item[['Purchase Count', 'Item Price', 'Total Purchase Value']]
popular_item.head()
```
